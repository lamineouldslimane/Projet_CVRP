<h4 style="color:#5c88d9"> 1.1.4 Problème du VRP : </h4>
Le problème de routage de véhicules (VRP : Vehicle Routing Problem) aussi appelé Problème de Tournées de Véhicules (PTV) est un problème dans lequel un certain nombre de clients doit être visités par un certain nombre de véhicules. Ces tournées doivent être organisées de façon à optimiser un certain nombre de critères tels que par exemple, la distance totale parcourue et les temps d'attente des clients. 
Il y a de nombreuses variantes de ce problème intégrant une ou plusieurs des contraintes suivantes : capacité des véhicules, périodes de temps pour visiter les clients, limites technologiques du système ou des situations géographiques des clients. 
Le problème de base de tournées de véhicules peut se définir ainsi :
Etant donné un graphe dont les nœuds représentent les clients et les arcs représentent les routes, le problème consiste à concevoir des itinéraires pour un ensemble des véhicules disponibles, afin d'une part de respecter les contraintes et d'autre part d'optimiser une fonction objectif donnée. 

<h4 style="color:#5c88d9"> 1.1.5  Dans notre cas le problème du CVRP : </h4>
Dans notre cas notre c’est porté sur le CVRP qui a pour extension VRPPD, VRPTW et DCVRP.
Le CVRP (Capacitated Vehicle Routing Problem) ou VRP avec contrainte de capacité est un ensemble de véhicules de capacité finie, basée dans un dépôt, qui se doit d’assurer des tournées entre plusieurs clients (ou villes) ayant demandé chacun une certaine quantité de marchandises. L'ensemble des clients visités par un véhicule désigne la tournée de celui-ci. Chaque client doit être desservi une et une seule fois et chaque tournée commence et se termine au dépôt.
L'objectif du CVRP est de minimiser le coût total, c-à-d la somme des distances ou des temps de parcours des tournées, tout en respectant la contrainte de capacité des véhicules : la quantité de marchandises livrées sur une tournée ne doit pas dépasser la capacité du véhicule qui l'assure.

<h3 style="color:#5c88d9"> 2.2 Étude de complexité : </h3>

La complexité problématique est relative au problème à résoudre ainsi que la méthode de résolution adoptée pour élaborer la solution optimale par rapport au critère retenu.
- Un problème de décision comprend deux parties : une partie donnée du problème et un processus binaire ayant « oui » ou « non » comme réponse possible.
- Un problème de recherche est un problème constitué d’un ensemble de données dont chacun représente un ensemble de solutions. Donc, la résolution d’un problème de recherche consiste à trouver pour chaque ensemble de données D des solutions S associées. 
- Un problème d’optimisation est un problème de recherche en associant à chaque solution une valeur qualitative. A chaque problème d’optimisation, on peut associer un problème de décision donc l’étude de la complexité du problème de décision peut donner des indications au problème d’optimisation associé. La théorie de la complexité permet de classer les problèmes en deux classes P et NP.
- La classe P regroupe les problèmes qui peuvent être résolus par des algorithmes polynomiaux. Un algorithme est dit polynomial, lorsque son temps d’exécution est borné par O(P(x)) ou p est un polynôme et x est la longueur d’entrée d’une instance du problème.
- Les algorithmes dont la complexité ne peut pas être bornée polynomial sont qualifiés d’exponentiels et correspondent à la classe NP.
- Un problème de décision est dit NP-complet s’il appartient à la classe NP et il résolu, au mieux, en un temps exponentiel.
- Un problème d’optimisation est dit NP-Difficile, si le problème de décision associé est
NP-complet.

<br>

Pour connaitre le problème de complexité du VRP, commençons tout d’abord par le problème du voyageur de commerce étant donné que le VRP est une extension du TSP avec certain paramètre en plus (le nombre de véhicules en plus, La capacité de chaque véhicule...). 
<h4 style="color:#5c88d9"> 2.2.1 Problème du voyageur de commerce (TSP)  : </h4>
Le TSP (Traveling Salesman Problem) consiste en la détermination d'un parcours de coût minimal (distance, temps, etc.) pour un seul véhicule partant d'une localité, visitant n autres localités et revenant à son point de départ.
En tant que problème d'optimisation, le TSP est un problème NP-difficile. En effet, dans le cas où le graphe associé n'est pas orienté, le nombre total de solutions possibles est ((n-1) !) /2 où n est le nombre de villes. Avec une telle complexité factorielle, une résolution efficace du TSP nécessite donc le recours à des heuristiques spécialisées voire même à des métaheuristiques.
Dans un sens, VRP est une extension de TSP car il ajoute des contraintes supplémentaires comme le nombre de véhicule ou la capacité du VRP, de ce fait par réductibilité. Nous pouvons on déduire que le TSP est au moins aussi difficile que le VRP.
<h4 style="color:#5c88d9"> 2.2.2 Complexité du VRP / CVRP  : </h4>
Nous pouvons donc en conclure comme « Lenstra » et « Rinnooy Kan » que le problème d'élaboration de tournées de véhicules est un problème NP-difficile, c'est-à-dire qu'il n'existe pas à ce jour un algorithme déterministe pouvant résoudre ce problème en temps polynomial.
Ce qui signifie que l'effort de calcul nécessaire pour résoudre ce problème augmente de manière exponentielle avec la taille du problème. Pour de tels problèmes, il est souvent souhaitable d'obtenir des solutions approximatives, afin qu'elles puissent être trouvées assez rapidement et soient suffisamment précises pour l'objectif visé.
 Généralement, cette tâche est accomplie en utilisant diverses méthodes heuristiques, qui reposent sur une certaine connaissance de la nature du problème.
Dans notre cas le CVRP est une variante du VRP avec une contrainte de capacité de ce fait-il est aussi NP-Difficile.

<h1 style="color:#4476d3"> 3. Artificial bee colony algorithm </h1>
<h3 style="color:#5c88d9"> 3.1 Le Concept : </h3>

L'algorithme ABC appartient à une classe d'algorithmes évolutionnaires qui s'inspirent de l'intelligence
comportement des abeilles mellifères dans la recherche de sources de nectar autour de leurs ruches. Cette classe de métaheuristique n'a commencé à retenir l'attention que récemment. Différentes variantes d'algorithmes d'abeilles sous divers
des noms ont été proposés pour résoudre des problèmes combinatoires. Mais dans tous, certains
des stratégies de recherche sont appliquées; c'est-à-dire que les solutions complètes ou partielles sont considérées comme des sources de nourriture
et les groupes d'abeilles tentent d'exploiter les sources de nourriture dans l'espoir de trouver du nectar de bonne qualité (c.-à-d.
solutions de haute qualité) pour la ruche. De plus, les abeilles communiquent entre elles sur la
fouiller l'espace et les sources de nourriture en exécutant une danse de waggle.<br>
Dans l'algorithme ABC, les abeilles sont divisées en trois types; abeilles employées, badauds et éclaireurs.<br>
Les abeilles employées sont responsables de l'exploitation des sources de nourriture disponibles et de la collecte requise
information. Ils partagent également les informations avec les spectateurs, et les spectateurs sélectionnent les
les sources de nourriture à explorer davantage. Lorsque la source de nourriture est abandonnée par son abeille employée,
l'abeille employée devient éclaireuse et commence à chercher une nouvelle source de nourriture à proximité de la ruche.<br>
L'abandon se produit lorsque la qualité de la source de nourriture n'est pas améliorée après avoir effectué une
nombre maximal d'itérations autorisé.<br><br>

L'algorithme ABC est un algorithme itératif, et il commence par générer des solutions aléatoires comme
sources de nourriture et affectation de chaque abeille employée à une source de nourriture. Puis à chaque itération, chaque
une abeille employée trouve une nouvelle source de nourriture près de sa source de nourriture (ou ancienne)
opérateur de quartier). La quantité de nectar (fitness) de la nouvelle source de nourriture est ensuite évaluée. Si la
la nouvelle source de nourriture contient plus de nectar que l'ancienne, puis l'ancienne est remplacée par la nouvelle.
Une fois que toutes les abeilles employées ont terminé le processus d'exploitation ci-dessus, elles partagent le nectar
information des sources de nourriture avec les badauds. Ensuite, chaque spectateur sélectionne une source de nourriture
selon la méthode traditionnelle de sélection des roues de roulette. Après cela, chaque spectateur trouve un aliment
source près de sa source de nourriture sélectionnée (en utilisant un opérateur de quartier) et calcule le nectar
quantité de nourriture voisine. Ensuite, pour chaque ancienne source de nourriture, la meilleure source de nourriture parmi tous
les sources de nourriture près de l'ancienne source de nourriture sont déterminées. L'abeille employée associée à l'ancien
la source de nourriture est affectée à la meilleure source de nourriture et abandonne l'ancienne si la meilleure source de nourriture est
mieux que l'ancienne source de nourriture. Une source de nourriture est également abandonnée par une abeille employée si la qualité
de la source de nourriture n'a pas été améliorée pour limiter (un nombre prédéterminé) les itérations successives.
Cette abeille employée devient alors éclaireuse et cherche au hasard une nouvelle source de nourriture. Après le
le scout trouve une nouvelle source de nourriture, le scout redevient une abeille employée. Après chaque abeille employée est affecté à une source de nourriture, une autre itération de l'algorithme ABC commence. L'ensemble du processus est
répété jusqu'à ce qu'une condition d'arrêt soit remplie.
